```python
# ==========================================
# STEP 1: INSTALL & IMPORT TOOLS
# ==========================================
!pip install tf-explain -q

import tensorflow as tf
from tf_explain.core.grad_cam import GradCAM
import matplotlib.pyplot as plt
import numpy as np
import cv2

# ==========================================
# STEP 2: THE MEDICAL IMAGING DIAGNOSIS (XAI)
# ==========================================
# We use VGG16, a model pre-trained by researchers to see images
model = tf.keras.applications.VGG16(weights="imagenet", include_top=True)

def run_medical_diagnosis(img_url):
    print("--- Starting Medical Image Diagnosis ---")

    # 1. Download the image
    img_path = tf.keras.utils.get_file("medical_scan.jpg", img_url)
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)

    # 2. Apply Grad-CAM (The "Explainable" part)
    # This creates a heatmap showing what the AI is looking at
    explainer = GradCAM()
    # 285 is a general class for 'chest/monitor' in this demo model
    # Explicitly specify the target layer for Grad-CAM, as automatic inference failed.
    # 'block5_conv3' is the last convolutional layer in VGG16 before flattening.
    heatmap = explainer.explain(([img_array], None), model, class_index=285, layer_name='block5_conv3')

    # 3. Display Results
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Original Medical Scan")
    plt.imshow(img)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("XAI Explanation (Heatmap)")
    plt.imshow(heatmap)
    plt.axis('off')
    plt.show()
    print("Diagnosis Complete: Regions of interest highlighted in Red/Yellow.")

# ==========================================
# STEP 3: THE DAILY HELP AI SYSTEM
# ==========================================
def daily_health_assistant(heart_rate, temp):
    print("\n--- Daily Health Assistant Report ---")
    status = "Normal"
    explanation = "All vitals are within the standard range for your age."

    if heart_rate > 100:
        status = "Warning: High Heart Rate"
        explanation = f"Your heart rate is {heart_rate} bpm. This is high because you are currently stationary but your pulse matches light exercise levels."

    if temp > 99.5:
        status = "Alert: Fever Detected"
        explanation = f"Your temperature is {temp}F. AI recommends rest and hydration."

    print(f"STATUS: {status}")
    print(f"EXPLANATION: {explanation}")

# ==========================================
# STEP 4: EXECUTE EVERYTHING
# ==========================================
# Testing with a sample Chest X-ray link
sample_xray = "https://raw.githubusercontent.com/ieee8023/covid-chestxray-dataset/master/images/000001-1.jpg"
run_medical_diagnosis(sample_xray)

# Testing the Daily Help System with sample data (e.g., Heart Rate 110, Temp 98.6)
daily_health_assistant(110, 98.6)
```

#### 7. Calibration Plot / Reliability Diagram (New Plot 1 of 2)

A calibration plot, or reliability diagram, assesses how well a probabilistic classification model is calibrated. A well-calibrated model's predicted probabilities should match the true event frequencies. For example, among the samples to which the model assigned a probability of 0.8, approximately 80% should actually belong to the positive class.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.calibration import calibration_curve
from sklearn.preprocessing import LabelBinarizer

# --- Generate Synthetic Data for Model Performance Metrics (for self-containment) ---
np.random.seed(42)
num_samples = 200
num_classes = 3
class_names = ['Class A', 'Class B', 'Class C']
y_true = np.random.randint(0, num_classes, num_samples)
y_pred_proba = np.random.rand(num_samples, num_classes)
y_pred_proba = y_pred_proba / y_pred_proba.sum(axis=1, keepdims=True)
y_pred = np.argmax(y_pred_proba, axis=1)

plt.figure(figsize=(10, 8))

# Binarize the true labels for multi-class plots
label_binarizer = LabelBinarizer()
y_true_binarized = label_binarizer.fit_transform(y_true)

for i in range(num_classes):
    prob_pos = y_pred_proba[:, i]
    fraction_of_positives, mean_predicted_value = calibration_curve(y_true_binarized[:, i], prob_pos, n_bins=10)
    plt.plot(mean_predicted_value, fraction_of_positives, "s-", label=f'{class_names[i]}')

plt.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")
plt.xlabel("Mean predicted value")
plt.ylabel("Fraction of positives")
plt.title("Calibration Plot (Reliability Diagram)")
plt.legend(loc="lower right")
plt.grid(True)
plt.savefig('calibration_plot.png')
plt.show()
```

#### 8. Predicted Probability Distribution (New Plot 2 of 2)

These histograms visualize the distribution of predicted probabilities for each class. This can help understand the model's confidence levels and whether it's making clear distinctions between classes or if probabilities are clustered around certain values.

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Generate Synthetic Data for Model Performance Metrics (for self-containment) ---
np.random.seed(42)
num_samples = 200
num_classes = 3
class_names = ['Class A', 'Class B', 'Class C']
y_true = np.random.randint(0, num_classes, num_samples)
y_pred_proba = np.random.rand(num_samples, num_classes)
y_pred_proba = y_pred_proba / y_pred_proba.sum(axis=1, keepdims=True)
y_pred = np.argmax(y_pred_proba, axis=1)

plt.figure(figsize=(12, 6))
for i in range(num_classes):
    plt.hist(y_pred_proba[:, i], bins=30, alpha=0.5, label=f'Class {class_names[i]}')
plt.title('Distribution of Predicted Probabilities for Each Class')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(True)
plt.savefig('predicted_probability_distribution.png')
plt.show()
```

#### 3. Classification Report Visualization (New Plot 1 of 4)

This bar chart visualizes the precision, recall, and F1-score for each class, giving a quick overview of how well the model performs for different categories.

```python
# 3. Classification Report Visualization
report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
df_report = pd.DataFrame(report).transpose().iloc[:-3]

plt.figure(figsize=(10, 6))
sns.heatmap(df_report.iloc[:, :3], annot=True, cmap='Blues', fmt=".2f", cbar=True)
plt.title('Classification Report Heatmap (Precision, Recall, F1-Score)')
plt.ylabel('Classes')
plt.xlabel('Metrics')
plt.tight_layout()
plt.savefig('classification_report_heatmap.png')
plt.show()
```

#### 4. Confusion Matrix (New Plot 2 of 4)

The confusion matrix provides a detailed breakdown of correct and incorrect classifications for each class, helping to identify specific areas where the model might be struggling.

```python
# 4. Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.savefig('confusion_matrix.png')
plt.show()
```

#### 5. ROC Curve (New Plot 3 of 4)

An ROC curve illustrates the diagnostic ability of a binary classifier system as its discrimination threshold is varied. It plots the True Positive Rate (Sensitivity) against the False Positive Rate (1-Specificity) at various threshold settings.

```python
# 5. ROC Curve (for multi-class, using One-vs-Rest strategy)

plt.figure(figsize=(10, 8))

# Binarize the true labels for multi-class ROC
label_binarizer = LabelBinarizer()
y_true_binarized = label_binarizer.fit_transform(y_true)

for i in range(num_classes):
    fpr, tpr, _ = roc_curve(y_true_binarized[:, i], y_pred_proba[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC curve of {class_names[i]} (area = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig('roc_curve.png')
plt.show()
```

#### 6. Precision-Recall Curve (New Plot 4 of 4)

The Precision-Recall curve is particularly useful for imbalanced datasets, as it highlights the trade-off between precision (the proportion of positive identifications that were actually correct) and recall (the proportion of actual positives that were correctly identified) at different classification thresholds.

```python
# 6. Precision-Recall Curve (for multi-class, using One-vs-Rest strategy)

plt.figure(figsize=(10, 8))

for i in range(num_classes):
    precision, recall, _ = precision_recall_curve(y_true_binarized[:, i], y_pred_proba[:, i])
    plt.plot(recall, precision, label=f'Precision-Recall curve of {class_names[i]}')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid(True)
plt.savefig('precision_recall_curve.png')
plt.show()
```
